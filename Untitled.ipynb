```python
# Every enmp entry is u16
entrylen = 2
entries = [
    "id",
    "level",
    "health",
    "maxDamage",
    "minDamage",
    "physicalWeakness",
    "fireWeakness",
    "iceWeakness",
    "thunderWeakness",
    "darkWeakness",
    "lightWeakness",
    "generalWeakness",
    "experience",
    "prize",
    "bonusLevel"
]
def toBytes(n):
    if n != 0:
        print(n)
    return n.to_bytes(2, "little")
    n_sig = n // 255
    n_insig = n % 255
    return bytearray([n_insig,n_sig])
import os, yaml
header = [2, 0, 0, 0, 229, 0, 0, 0] #taken from vanilla file
enmp = os.path.join("khbr","data","enmpVanilla.yml")
enmplist = yaml.load(open(enmp))
enmp_bytes = bytearray(header)
for enemy in enmplist:
    for k in entries:
        if k == "health":
            for healthentry in enemy[k]:
                enmp_bytes += toBytes(healthentry)
            continue
        enmp_bytes += toBytes(enemy[k])
vanillabytes = bytearray(open("C:\\Users\\Arcade\\Desktop\\pcsx2-ex\\KH2\\bars\\root\\00battle\\enmp.list", "rb").read())
vanillabytes = vanillabytes
# Write the two files out in non byte form in rows of 16 that are easy to diff
assert len(vanillabytes) == len(enmp_bytes), (len(vanillabytes), len(enmp_bytes))
with open("test-avanilla", "w") as f:
    with open("test-bnew", "w") as f_n:
        f.write(" ".join([str(b) for b in vanillabytes[:8]]))
        f_n.write(" ".join([str(b) for b in enmp_bytes[:8]]))
        
        for _ in range(9,len(enmp_bytes), 16):
            print("ummm")
            f.write("\n")
            f_n.write("\n")
            
            f.write(" ".join([str(b) for b in vanillabytes[_:_+16]]))
            f_n.write(" ".join([str(b) for b in enmp_bytes[_:_+16]]))
assert vanillabytes == enmp_bytes
```
